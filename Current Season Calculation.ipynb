```python
%matplotlib inline

import pandas as pd
import numpy as np
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
file_path = "premier_league_table_23.csv"

# Load the latest version
df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "siddhrajthakor/premier-league-table-202526",
  file_path,
)[["Team", "Played", "Goals For", "Goals Against", "Points"]]
```

```python
k = 1.2095798373742814
c = np.exp(-0.16786731341479072)
```

```python
currentSeason = df.rename(columns={
    "Goals For": "GF",
    "Goals Against": "GA",
    "Played": "G",
    "Points": "P"
})

currentSeason["xP"] = (c * currentSeason["GF"]**k / (c * currentSeason["GF"]**k + currentSeason["GA"]**k)) * 3 * currentSeason["G"]
currentSeason["Error"] = currentSeason["P"] - currentSeason["xP"]
currentSeason = currentSeason.sort_values("xP", ascending=False)
```

```python
print("\nThe Expected Table for the Current 2025-2026 Season\n")
display(currentSeason.style.hide(axis="index"))
```
